```python
import tensorflow as tf
from tensorflow import keras
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train[1000]
```

```python
y_train[1000]
```

```python
x_train=x_train/255.0 #normailization
x_test=x_test/255.0
```

```python
print(x_train[0])
```

```python
#buiding neural network
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)), #convert into one dimension data 28+28=784 [para 0 no calculations are performed 
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(10,activation='softmax')
])
model.summary()
```

```python
#compile
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
#train model
model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
#evaluate
model.evaluate(x_test,y_test)
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
print("loss=",loss)
print("accuracy=",accuracy)
```

```python
import numpy as np
sample_image=x_test[0]
sample_label=y_test[0]
sample_image
```

```python
#add batch dimensions
sample_image=np.expand_dims(sample_image,axis=0)
```

```python
predictions=model.predict(sample_image)
predicted_digit=np.argmax(predictions)
print("prediction digit:",predicted_digit)
print("actual digit:",sample_label)
```
