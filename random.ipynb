```python
import numpy as np
import pandas as pd
```

```python

np.random.seed(42)
n = 200

experience = np.random.randint(0, 10, n)
education = np.random.randint(1, 4, n)  # 1=UG, 2=PG, 3=PhD
skill_score = np.random.randint(1, 11, n)

# Salary formula (realistic logic + noise)
salary = (
    experience * 5000 +
    education * 8000 +
    skill_score * 3000 +
    np.random.normal(0, 5000, n)
)

data = pd.DataFrame({
    'experience': experience,
    'education': education,
    'skill_score': skill_score,
    'salary': salary
})

data.head()
```

```python
x=data[['experience','education','skill_score']]
y=data['salary']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
model=Sequential([#empty model
   Dense(16,activation="relu",input_shape=(3,)),
   Dropout(0.5),
   Dense(8,activation="relu"),
   Dropout(0.5),
   Dense(4,activation="relu"),
   Dropout(0.5),
   Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer="adam",loss="mean_squared_error",metrics=["mae"])
```

```python
model.fit(x,y,epochs=20)
```

```python
loss,mae=model.evaluate(x_test,y_test)
print("loss",loss)
print("mae",mae)
```
